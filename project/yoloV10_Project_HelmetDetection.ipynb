```python
YAML_PATH = '/content/safety_helmet_dataset/data.yaml'
```

```python
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 32
```

```python
model.train(data=YAML_PATH, epochs=EPOCHS, imgsz=IMG_SIZE, batch=BATCH_SIZE)
```

```python
TRAINED_MODEL_PATH = '/content/yolov10/runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)
model.val(data=YAML_PATH,imgsz=IMG_SIZE,split='test')
```

```python
#TEST
img_path1 = '/content/safety_helmet_dataset/test/images/helmet-10-_jpg.rf.eee985865a1a52407316f0c9ea7a9b3e.jpg'
from IPython.display import Image

# Display the image
Image(img_path1)
```
